```python
from random import choice
from Bio import SeqUtils
from Bio.SeqUtils import MeltingTemp as mt
Nucelotides = 'ACGT'
Length = 30
min_GC = 45
max_GC = 60
min_Tm = 50
max_Tm = 60
Oligo_List = []
while len(Oligo_List)<10:
    candidate = ''.join(choice(Nucelotides) for i in range(Length))
    Gc_cont = SeqUtils.GC(candidate)
    if Gc_cont < min_GC or Gc_cont > max_GC :
        continue
    Tm = mt.Tm_Wallace(candidate)
    if Tm < min_Tm or Tm > max_Tm :
        continue
    Oligo_List.append(candidate)
    print(candidate,int(Gc_cont),int(Tm))
```

```python
import pandas as pd
f = '/bigstore/binfo/24bit_adapter_pairs.fasta'
test = pd.read_csv(f,sep='\n',header=None)
test
```

```python
Adapter = []
Sequence = []
i
for i in range(0,len(test),2):
    if i%10000 ==0:
        print(i)
    Adapter.append(test.iloc[i][0].split('>')[1])
    Sequence.append(test.iloc[i+1][0])
```

```python
Adapters = pd.DataFrame(columns = ['name','sequence'])
Adapters['name'] = Adapter
Adapters['sequence'] = Sequence
```
