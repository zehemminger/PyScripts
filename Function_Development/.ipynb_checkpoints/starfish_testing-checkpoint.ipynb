```python
import matplotlib.pyplot as plt
import numpy as np
from copy import deepcopy
from starfish import data, FieldOfView, display
from starfish.image import Filter
from starfish.spots import DetectPixels
from starfish.types import Axes, Features, Levels

# Load MERFISH data
experiment = data.MERFISH(use_test_data=True)
imgs = experiment.fov().get_image(FieldOfView.PRIMARY_IMAGES)

# filter and deconvolve data
ghp = Filter.GaussianHighPass(sigma=3)
dpsf = Filter.DeconvolvePSF(num_iter=15, sigma=2, level_method=Levels.SCALE_SATURATED_BY_CHUNK)
glp = Filter.GaussianLowPass(sigma=1)
ghp.run(imgs, in_place=True)
dpsf.run(imgs, in_place=True)
glp.run(imgs, in_place=True)

# scale data with user-defined factors to normalize images. For this data set, the scale factors
# are stored in experiment.json.
scale_factors = {
    (t[Axes.ROUND], t[Axes.CH]): t['scale_factor']
    for t in experiment.extras['scale_factors']
}
filtered_imgs = deepcopy(imgs)
for selector in imgs._iter_axes():
    data = filtered_imgs.get_slice(selector)[0]
    scaled = data / scale_factors[selector[Axes.ROUND.value], selector[Axes.CH.value]]
    filtered_imgs.set_slice(selector, scaled, [Axes.ZPLANE])

# Decode with PixelSpotDecoder
psd = DetectPixels.PixelSpotDecoder(
    codebook=experiment.codebook,
    metric='euclidean',             # distance metric to use for computing distance between a pixel vector and a codeword
    norm_order=2,                   # the L_n norm is taken of each pixel vector and codeword before computing the distance. this is n
    distance_threshold=0.5176,      # minimum distance between a pixel vector and a codeword for it to be called as a gene
    magnitude_threshold=1.77e-5,    # discard any pixel vectors below this magnitude
    min_area=2,                     # do not call a 'spot' if it's area is below this threshold (measured in pixels)
    max_area=np.inf,                # do not call a 'spot' if it's area is above this threshold (measured in pixels)
)
initial_spot_intensities, prop_results = psd.run(filtered_imgs)

# filter spots that do not pass thresholds
spot_intensities = initial_spot_intensities.loc[initial_spot_intensities[Features.PASSES_THRESHOLDS]]

# Example of how to access the spot attributes
print(f"The area of the first spot is {prop_results.region_properties[0].area}")

# View labeled image after connected componenet analysis
plt.imshow(prop_results.label_image[0])
plt.title("PixelSpotDecoder Labeled Image")

# View decoded spots overlaid on max intensity projected image
single_plane_max = filtered_imgs.reduce({Axes.ROUND, Axes.CH, Axes.ZPLANE}, func="max")
# Uncomment code below to view spots
#%gui qt
#viewer = display(stack=single_plane_max, spots=spot_intensities)
```

```python
fishdata_path = '/hybedata/Images/Zach/ZebraFinch/A4_2020Jun28/fishdata/'
from MERFISH_Objects.FISHData import *
fishdata = FISHData(fishdata_path)
import os
posname = 'Pos1'
from analysis_scripts.classify import *
cword_config = 'seqfish_config_zebrafinch'
seqfish_config = importlib.import_module(cword_config)
try:
    genes = seqfish_config.gids+seqfish_config.bids
except:
    genes = seqfish_config.gids
bitmap = seqfish_config.bitmap
normalized_gene_vectors = seqfish_config.norm_gene_codeword_vectors
normalized_all_gene_vectors = seqfish_config.norm_all_codeword_vectors
all_gene_vectors = seqfish_config.all_codeword_vectors
nbits = seqfish_config.nbits
```

```python
from tqdm import tqdm
import hashlib
dataset = 'A4_2020Jun28'
channels = list(np.unique([channel for seq,hybe,channel in bitmap]))
channels.sort()
hybes = list(np.unique([hybe for seq,hybe,channel in bitmap]))
hybes.sort()
channel_indexes = {channel:i for i,channel in enumerate(channels)}
hybe_indexes = {hybe:i for i,hybe in enumerate(hybes)}
acqs = [i for i in metadata.image_table.acq.unique() if 'hybe' in i]
posnames = metadata.image_table[metadata.image_table.acq.isin(acqs)].Position.unique()
posnames.sort()
posname_indexes = {posname:i for i,posname in enumerate(posnames)}
imaged_zindexes = metadata.image_table[metadata.image_table.Position.isin(posnames)].Zindex.unique()
zindexes = list(range(np.min(imaged_zindexes)+seqfish_config.parameters['projection_k']-1,
                            np.max(imaged_zindexes)-seqfish_config.parameters['projection_k'],
                            seqfish_config.parameters['projection_zskip']))
zindexes.sort()
zindex_indexes = {zindex:i for i,zindex in enumerate(zindexes)}
n_colors = len(channels)
n_rounds = len(hybes)
n_zindexes = len(zindexes)
pixel_size = 0.109
z_step = 0.4
x_shape = 2048
y_shape = 2048
primary_images = {"version":"0.0.0",
                "contents":{},
                "extras":{}}

for posname in tqdm(posnames):
    posname_image_table = metadata.image_table[metadata.image_table.Position==posname]
    X,Y = posname_image_table.XY.iloc[0]
    Z = posname_image_table.Z.iloc[0]
    x_coordinates_min = X
    x_coordinates_max = X+(x_shape*pixel_size)
    y_coordinates_min = Y
    y_coordinates_max = Y+(y_shape*pixel_size)
    flag = fishdata.load_data('flag',dataset=dataset,posname=posname)
    if flag == 'Passed':
        for zindex in zindexes:
            tile_list = []
            exists = True
            z_index = zindex_indexes[zindex]
            Z = posname_image_table[posname_image_table.Zindex==zindex].Z.iloc[0]
            z_coordinates_min = Z
            z_coordinates_max = Z+z_step
            for hybe in hybes:
                if not exists:
                    continue
                r_index = hybe_indexes[hybe]
                for channel in channels:
                    if not exists:
                        continue
                    c_index = channel_indexes[channel]
                    fname = fishdata.generate_fname(dataset,posname,hybe,channel,zindex,'image')
                    if not os.path.exists(os.path.join(fishdata.base_path,fname)):
                        exists = False
                        continue
                    hasher = hashlib.sha256()
                    with open(os.path.join(fishdata.base_path,fname), 'rb') as fh:
                        hasher.update(fh.read())
                    sha256 = hasher.hexdigest()
                    tile = {"coordinates": {
                                    "xc": [
                                        x_coordinates_min,
                                        x_coordinates_max
                                    ],
                                    "yc": [
                                        y_coordinates_min,
                                        y_coordinates_max
                                    ],
                                    "zc": [
                                        z_coordinates_min,
                                        z_coordinates_max
                                    ]
                                },
                                "file": fname,
                                "indices": {
                                    "c": c_index,
                                    "r": r_index,
                                    "z": z_index
                                },
                                "sha256": sha256,
                                "tile_format": "TIFF",
                                "tile_shape": {
                                    "x": x_shape,
                                    "y": y_shape
                                }
                            }
                    tile_list.append(tile)
            fov_json = {"default_tile_format": "TIFF",
                    "dimensions": [
                        "z",
                        "xc",
                        "x",
                        "yc",
                        "y",
                        "zc",
                        "c",
                        "r"
                    ],
                    "extras": {},
                    "shape": {
                        "c": n_colors,
                        "r": n_rounds,
                        "z": n_zindexes
                    },
                    "tiles": tile_list,
                    "version": "0.1.0"
                }
            if exists:
                fov = "fov_"+str(posname_indexes[posname]).zfill(3)+str(zindex_indexes[zindex]).zfill(3)
                fov_json_name = "primary-images-{0}.json".format(fov)
                primary_images["contents"][fov]=fov_json_name
                out_file = open(os.path.join(fishdata_path,fov_json_name), "w")  
                json.dump(fov_json, out_file)  
                out_file.close()
    break
out_file = open(os.path.join(fishdata_path,"primary_images.json"), "w")  
json.dump(primary_images, out_file)  
out_file.close()
```

```python
mapping_list = []
codeword_vectors = (normalized_all_gene_vectors/0.5).astype('int16')
for gene_index in range(codeword_vectors.shape[0]):
    barcode = codeword_vectors[gene_index,:]
    codewords = []
    for bit_index in range(codeword_vectors.shape[1]):
        seq,hybe,channel = bitmap[bit_index]
        codeword = {"c":channel_indexes[channel],"r":hybe_indexes[hybe],"v":int(barcode[bit_index])}
        codewords.append(codeword)
    mapping = {"codeword":codewords,
           "target":genes[gene_index]}
    mapping_list.append(mapping)
codebook_json = {"version": "0.0.0",
                 "mappings": mapping_list}
# Save JSON
# print(codebook_json)
out_file = open(os.path.join(fishdata_path,"codebook.json"), "w")  
json.dump(codebook_json, out_file)  
out_file.close() 
```
