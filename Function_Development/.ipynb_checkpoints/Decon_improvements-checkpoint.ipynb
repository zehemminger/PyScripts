```python
from ipypb import ipb
"""
Implement a mean filter without the center
"""
def find_hot_pixels(md,n_pos=5,thresh=1.1,n_acqs=5,sigma=1):
    X = []
    Y = []
    n_imgs = 0
    poses = md.image_table.Position.unique()
    if len(poses)>n_pos:
        poses = random.sample(list(poses),n_pos)
    for pos in ipb(poses):
        pos_md =  md.image_table[md.image_table.Position==pos]
        acqs = pos_md.acq.unique()
        if len(acqs)>n_acqs:
            acqs = random.sample(list(acqs),n_acqs)
        for acq in acqs:
            hot_pixel_dict[acq] = {}
            channels = pos_md[pos_md.acq==acq].Channel.unique()
            for channel in channels:
                n_imgs+=1
                img = np.average(md.stkread(Position=pos,Channel=channel,acq=acq),axis=2)
                bkg_sub = img/gaussian(img,sigma=sigma)
                loc = np.where(bkg_sub>thresh)
                X.extend(loc[0])
                Y.extend(loc[1])
    print(n_imgs)
    hist = np.histogram2d(X,Y,bins=2048,range=[[0,2048],[0,2048]])
    loc = np.where(hist[0]>n_imgs/10)
    return hist[0],loc
img,loc = find_hot_pixels(md)
Display(img,rel_min=0,rel_max=100)
plt.scatter(loc[0],loc[1])
plt.show()
```
