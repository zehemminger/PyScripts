```python
from fish_helpers import *
plt.style.use(['dark_background'])
base = '/bigstore/binfo/mouse/Brain/'
from scipy.io import mmread
```

```python
datasets = {}
```

```python
method = 'FISH'
dataset = 'seqFISH_Neuron2016_Hippocampus1'
datasets[dataset] = {}
datasets[dataset]['cells'] = pd.read_csv(os.path.join(base,method,dataset,'cells.csv'),index_col=0)
datasets[dataset]['genes'] = pd.read_csv(os.path.join(base,method,dataset,'genes.csv'),index_col=0)
datasets[dataset]['metadata'] = pd.read_csv(os.path.join(base,method,dataset,'metadata.csv'),index_col=0)
datasets[dataset]['counts'] = pd.read_csv(os.path.join(base,method,dataset,'counts.csv'),index_col=0)
```

```python
method = 'FISH'
dataset = 'seqFISH_Neuron2016_Hippocampus2'
datasets[dataset] = {}
datasets[dataset]['cells'] = pd.read_csv(os.path.join(base,method,dataset,'cells.csv'),index_col=0)
datasets[dataset]['genes'] = pd.read_csv(os.path.join(base,method,dataset,'genes.csv'),index_col=0)
datasets[dataset]['metadata'] = pd.read_csv(os.path.join(base,method,dataset,'metadata.csv'),index_col=0)
datasets[dataset]['counts'] = pd.read_csv(os.path.join(base,method,dataset,'counts.csv'),index_col=0)
```

```python
method = 'Sequencing'
dataset = 'Smartseq'
datasets[dataset] = {}
datasets[dataset]['cells'] = pd.read_csv(os.path.join(base,method,dataset,'Hippocampus','cells.csv'),index_col=0)
datasets[dataset]['genes'] = pd.read_csv(os.path.join(base,method,dataset,'Hippocampus','genes.csv'),index_col=0)
datasets[dataset]['metadata'] = pd.read_csv(os.path.join(base,method,dataset,'Hippocampus','metadata.csv'),index_col=0)
counts = mmread(os.path.join(base,method,dataset,'Hippocampus','counts.mtx'))
datasets[dataset]['counts'] = pd.DataFrame(counts.A,index=list(datasets[dataset]['cells']['x']),columns=list(datasets[dataset]['genes']['x']))
```

```python
method = 'FISH'
dataset = 'seqFISH_Neuron2016_Hippocampus1'
pd.DataFrame(index=list(datasets[dataset]['counts'].columns)).to_csv(os.path.join(base,method,dataset,'genes.csv'))
```
