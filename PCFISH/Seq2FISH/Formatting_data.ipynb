```python
from scipy.io import mmread
method = 'Sequencing'
dataset = '10X_cells_v2_AIBS'
data = mmread(os.path.join(base,method,dataset,'Source','matrix.mtx.gz'))
data
```

```python
data.shape
```

```python
features = pd.read_csv(os.path.join(base,method,dataset,'Source','features.tsv'),sep='\t',header=None)
genes = pd.DataFrame(index = list(features[1]))
barcode = pd.read_csv(os.path.join(base,method,dataset,'Source','barcode.tsv'))
cells = pd.DataFrame(index = list(barcode.x))
```

```python
metadata = pd.DataFrame(list(cells),columns = ['Cell_ID'])
meta = pd.read_csv(os.path.join(base,method,dataset,'Source','sample_metadata.csv'),index_col=0)
metadata[meta.columns] = meta
metadata
```

```python
classes = pd.read_csv(os.path.join(base,method,dataset,'Source','cluster.membership.csv'),index_col=0)
metadata['cluster_index'] = classes
metadata
```

```python
cluster_annotations = pd.read_csv(os.path.join(base,method,dataset,'Source','cluster.annotation.csv'),index_col=0)
temps = []
for cluster in cluster_annotations.index:
    temp = pd.DataFrame(index = metadata[metadata.cluster_index==cluster].index,columns = list(cluster_annotations.columns))
    for column in list(cluster_annotations.columns):
        temp[column] = cluster_annotations[column].loc[cluster]
    temps.append(temp)
metadata[cluster_annotations.columns] = pd.concat(temps)
metadata
```

```python
genes.to_csv(os.path.join(base,method,dataset,'genes.csv'))
cells.to_csv(os.path.join(base,method,dataset,'cells.csv'))
metadata.to_csv(os.path.join(base,method,dataset,'metadata.csv'))
```
