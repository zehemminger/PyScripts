```python
fpkm_path = '/bigstore/binfo/zebra_finch/RegionXExpression/cufflinks_07Oct2019/FPKM_NCBI_Appended.csv'
fpkm = pd.read_csv(fpkm_path)
fpkm = fpkm[fpkm.gene_short_name.isin(spotcalls.gene.unique())]
fpkm
```

```python
from scipy.stats import pearsonr
F = []
S = []
G = []
df = spotcalls
for gene,cc in Counter(df.gene).items():
    G.append(gene)
    F.append(cc)
    try:
        s = np.max(fpkm[fpkm.gene_short_name==gene].FPKM)
        if isinstance(s,float):
            S.append(np.max(fpkm[fpkm.gene_short_name==gene].FPKM))
        else:
            S.append(np.nan)
    except:
        S.append(np.nan)
print(spearmanr(S,F))
print(pearsonr(S,F))
plt.scatter(np.log10(np.array(S)+1),np.log10(np.array(F)+1))
plt.show()
```

```python
from scipy.ndimage import gaussian_filter
x = np.linspace(np.log10(np.percentile(spotcalls.ave,1)),np.log10(np.percentile(spotcalls.ave,99)),num=100)
y = np.linspace(np.log10(np.percentile(spotcalls.cword_dist,1)),np.log10(np.percentile(spotcalls.cword_dist,99)),num=100)
plt.figure(figsize=[10,10])
img = np.histogram2d(np.log10(spotcalls.ave),np.log10(spotcalls.cword_dist),bins=[x,y])[0].T
plt.imshow(np.log10(gaussian_filter(img,2)+1),cmap='inferno')
t = (np.linspace(0,99,num=5)).astype(int)
plt.xticks(ticks = t,labels = [round(i,3) for i in x[t]])
plt.yticks(ticks = t,labels = [round(i,3) for i in 10**y[t]])
plt.xlabel('log10 Average Intensity')
plt.ylabel('Cword Distance')
plt.colorbar()
plt.show()
```
