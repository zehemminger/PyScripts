```python
def merfish_correlation(spotcalls,color='b',alpha=1,label='spotcalls'):
    #ReadsPerGene = pd.read_excel('/bigstore/GeneralStorage/Zach/Cornea_RNAseq/Aligned/ReadsPerGene.xlsx')
    f = '/bigstore/GeneralStorage/Evan/NFKB_MERFISH/Calibration_Set_Data/3T3_Calibration_Set/RNA_Seq_Data/Hoffmann_IFNAR_KO_3T3_TNF.txt'
    hoffmann = pd.read_csv(f,sep='\t')
    gids = []
    for gid in hoffmann.Geneid:
        gids.append(gid.split('.')[0])
    hoffmann.Geneid=gids
    GeneList = pd.read_excel('/bigstore/GeneralStorage/Zach/MERFISH/Inflammatory/InflammationGeneList.xlsx')
    vals = []
    FISH_Spots = Counter(spotcalls.gene)
    for gn,cc in FISH_Spots.items():
        if 'blank' in gn:
            continue
        else:
            gid = GeneList.Gene_ID[GeneList.Gene==gn].tolist()[0]
            counts = hoffmann[hoffmann.Geneid==gid]['IFNAR-TNF-3_tot.bam'].values
            fpkm = counts/GeneList.Length[GeneList.Gene==gn].tolist()[0]
            if len(fpkm)>0:
                if cc<2:
                    continue
                vals.append((fpkm[0],cc))
    fpkms,ccs = zip(*vals)
    from scipy.stats import spearmanr
    plt.scatter(np.log2(fpkms),np.log2(ccs),c=color,alpha=alpha,label=label)
    print(spearmanr(fpkms,ccs))
    plt.suptitle('Untrimmed FPKM vs Spot Count')
    plt.ylabel('log2 MERFISH Spot Count')
    plt.xlabel('log2 RNAseq FPKM')
    plt.legend()
```

```python
def merfish_correlation(spotcalls,color='b',alpha=1,label='spotcalls',system='cornea'):
    if system=='cornea':
        f = '/bigstore/GeneralStorage/Zach/Cornea_RNAseq/Aligned/ReadsPerGene.xlsx'
        ReadsPerGene = pd.read_excel(f)
        ReadsPerGene.index = ReadsPerGene.GeneIDs
    elif system=='3t3':
        f = '/bigstore/GeneralStorage/Evan/NFKB_MERFISH/Calibration_Set_Data/3T3_Calibration_Set/RNA_Seq_Data/Hoffmann_IFNAR_KO_3T3_TNF.txt'
        hoffmann = pd.read_csv(f,sep='\t')
        gids = []
        for gid in hoffmann.Geneid:
            gids.append(gid.split('.')[0])
        hoffmann.Geneid = gids
        hoffmann.index = gids
    else:
        print('Unknown System')
    GeneList = pd.read_excel('/bigstore/GeneralStorage/Zach/MERFISH/Inflammatory/InflammationGeneList.xlsx')
    GeneList.index = GeneList.Gene
    counts = []
    fpkms = []
    FISH_Spots = Counter(spotcalls.gene)
    for gn,cc in FISH_Spots.items():
        if 'blank' in gn:
            continue
        else:
            gid = GeneList.loc[gn]['Gene_ID']
            if system=='cornea':
                reads = ReadsPerGene.loc[gid].Unstranded
            elif system=='3t3':
                reads = hoffmann.loc[gid]['IFNAR-TNF-3_tot.bam']
            else:
                print('Unknown System')
            fpkm = reads/GeneList.loc[gn]['Length']
            if isinstance(fpkm,np.float64):
                if cc<2:
                    continue
                counts.append(cc)
                fpkms.append(fpkm)
    from scipy.stats import spearmanr
    import matplotlib.pyplot as plt
    plt.scatter(np.log10(fpkms),np.log10(counts),c=color,alpha=alpha,label=label)
    print(spearmanr(fpkms,counts))
    plt.suptitle('Untrimmed FPKM vs Spot Count')
    plt.ylabel('log10 MERFISH Spot Count')
    plt.xlabel('log10 RNAseq FPKM')
    plt.legend()
```

# Segmenting cells and spliting groups

```python
#Segmenting Cells 
import numpy as np
import pandas as pd
from skimage.transform import resize
from skimage import filters
from skimage import morphology
from skimage.morphology import watershed
from scipy import ndimage


def segment_pos(pos_df,nuc_img):
    pos = pos_df.posname.iloc[0]
    xy = np.stack(pos_df.centroid,axis=0)
    hist2d = np.histogram2d(xy[:,0],xy[:,1],bins=200)
    up_hist2d =resize(hist2d[0],(2048,2048)) 

    nuc = nuc_img/np.percentile(nuc_img.ravel(),99)
    spots = up_hist2d/np.percentile(up_hist2d.ravel(),99)
    nuc_spots = filters.gaussian(nuc+spots, 10)

    binary = nuc_img>np.percentile(nuc_img.ravel(),90)
    binary = morphology.remove_small_objects(binary, 100)
    binary = morphology.remove_small_holes(binary, 100)
    labels = ndimage.label(binary)[0]

    mask = nuc_spots>np.percentile(nuc_spots.ravel(),60)
    mask = morphology.remove_small_objects(mask, 10000)
    mask = morphology.remove_small_holes(mask, 10000)
    mask = ndimage.binary_dilation(mask,iterations=20)

    segmented = watershed(1/nuc_spots,markers=labels,mask=mask)

    cell = []
    for x,y in pos_df.centroid:
        cell.append(str(pos+'_'+str(segmented[int(x),int(y)])))
    pos_df['cell']=cell
    print('\n',segmented.max(),' Cells found in ',pos)
    return pos_df

def segment_wrapper(df,md,nucacq=False,Zindex=15):
    if nucacq == False:
        nucacq = [i for i in md.image_table.acq.unique() if 'nucstain' in i][0]
    results = []
    for pos in df.posname.unique():
        pos_df = df[df.posname==pos]
        nuc_img = md.stkread(Position=pos,Channel='DeepBlue',acq=nucacq,Zindex=Zindex)[:,:,0]
        results.append(segment_pos(pos_df,nuc_img))
    segmented_df = pd.concat(results,ignore_index=True)
    return segmented_df
segmented_df = segment_wrapper(purged_df,md)
pickle.dump(segmented_df,open(os.path.join(md_path,'results','spotcalls_segmented.pkl'),'wb'))
```

```python
# adding stage coordinate to spotcalls
X = []
Y = []
pixelsize = 0.109
cameradirection = [1,1]
temp = spotcalls
for pos in temp.posname.unique():
    print(pos)
    CoordXY = md.image_table[md.image_table.Position==pos].XY.iloc[0]
    coordX = CoordXY[0]
    coordY = CoordXY[1]
    pos_temp = temp[temp.posname==pos]
    pos_centroid = pos_temp.centroid
    for xy in pos_centroid:
#         xy = pos_temp.centroid.iloc[i]
        rx = xy[1]-2048
        ry = xy[0]#-2048
        x = rx*pixelsize*cameradirection[0]+coordX
        y = ry*pixelsize*cameradirection[1]+coordY
        X.append(x)
        Y.append(y)
spotcalls['CoordX'] = X
spotcalls['CoordY'] = Y
```

## Looking now at Differential Expression

```python
# RNAseq from hoffmann
f = '/bigstore/GeneralStorage/Evan/NFKB_MERFISH/Calibration_Set_Data/3T3_Calibration_Set/RNA_Seq_Data/Hoffmann_IFNAR_KO_3T3_TNF.txt'
hoffmann = pd.read_csv(f,sep='\t')
gids = []
for gid in hoffmann.Geneid:
    gids.append(gid.split('.')[0])
hoffmann.Geneid=gids
GeneList = pd.read_excel('/bigstore/GeneralStorage/Zach/MERFISH/Inflammatory/InflammationGeneList.xlsx')
```

```python
X_counts = Counter(no_tnf_spotcalls.gene)
Y_counts = Counter(high_tnf_spotcalls.gene)
X_cells = len(no_tnf_spotcalls.cell.unique())
Y_cells = len(high_tnf_spotcalls.cell.unique())
no_vs_high = []
fpkms = []
for gene in X_counts.keys():
    if 'blank' in gene:
        continue
    else:
        Y = Y_counts[gene]/Y_cells
        X = X_counts[gene]/X_cells
        no_vs_high.append(Y/X)
        gid = GeneList.Gene_ID[GeneList.Gene==gene].tolist()[0]
        counts = hoffmann[hoffmann.Geneid==gid]['IFNAR-TNF-3_tot.bam'].values
        fpkms.append(counts/GeneList.Length[GeneList.Gene==gene].tolist()[0])
plt.scatter(np.log2(fpkms),np.log2(no_vs_high))
plt.xlabel('log2 RNAseq Expression (fpkm)')
plt.ylabel('log2 fold expression high/no \n normalized counts/cell')
plt.title('High TNF vs No TNF')
plt.show()
```

```python
ps = []
for gene in sc_df:
    ps.append(stats.ttest_ind(no_sc_df[gene],high_sc_df[gene]).pvalue)
ps_df = pd.DataFrame(data=ps,index=sc_df.columns)
ps_df = ps_df.sort_values(by=0,ascending=True)
```

```python
N_cells = len(no_tnf_spotcalls.cell.unique())
L_cells = len(low_tnf_spotcalls.cell.unique())
H_cells = len(high_tnf_spotcalls.cell.unique())
N=[]
L=[]
H=[]
NE = []
LE = []
HE = []
genes=[]

for gene in ps_df.index[0:17]:
    if not gene in no_sc_df.keys():
        continue
    if not gene in low_sc_df.keys():
        continue
    if not gene in high_sc_df.keys():
        continue
    n = no_sc_df[gene]
    l = low_sc_df[gene]
    h = high_sc_df[gene]
    ne = np.std(no_sc_df[gene])/N_cells
    le = np.std(low_sc_df[gene])/L_cells
    he = np.std(high_sc_df[gene])/H_cells
    N.append(np.sum(n)/N_cells)
    L.append(np.sum(l)/L_cells)
    H.append(np.sum(h)/H_cells)
    NE.append(ne)
    LE.append(le)
    HE.append(he)
    genes.append(gene)
indices = np.arange(len(genes))
width = 0.3
fig = plt.figure(figsize=[18,8])
ax = fig.add_subplot(111)
ax.bar(indices-width,N,width,label='No TNF',yerr=NE)
ax.bar(indices,L,width,label='Low TNF',yerr=LE)
ax.bar(indices+width,H,width,label='High TNF',yerr=HE)
ax.set_xticks(indices)
ax.set_xticklabels(genes)
ax.set_xlabel('Genes')
ax.set_ylabel('Counts/cell')
plt.title('Top 10% of differentially expressed genes by p value')
plt.legend()
plt.show()
```

#### Interpretation:
<br>After segmenting cells and trimming cells with low confidence and genes with low confidence
<br>I seperated the spotcalls into No,Low and High TNF datasets
<br>Normalizing the spot counts for each gene by the number of cells in each dataset
<br>Shown are the top 10% of genes by p value
<br>Some show dose dependance which is nice
<br>Looking at the genes all of the TNF genes are present (positive control)
<br>We also see a good number of NFKB related genes

```python
#Normalized by total expression for each cell (Aproximation of cell size)
no_sc_norm = np.array(no_sc_df)/np.sum(no_sc_df,axis=1)[:,None]
no_sc_norm_df = pd.DataFrame(data=no_sc_norm,columns=no_sc_df.columns,index=no_sc_df.index)
low_sc_norm = np.array(low_sc_df)/np.sum(low_sc_df,axis=1)[:,None]
low_sc_norm_df = pd.DataFrame(data=low_sc_norm,columns=low_sc_df.columns,index=low_sc_df.index)
high_sc_norm = np.array(high_sc_df)/np.sum(high_sc_df,axis=1)[:,None]
high_sc_norm_df = pd.DataFrame(data=high_sc_norm,columns=high_sc_df.columns,index=high_sc_df.index)

for i in [0,5,10,15,20]:
    plt.figure(figsize=[18,8])
    df = pd.DataFrame(index=sc_df.index)
    for gene in ps_df.index[i:i+4]:
        no = np.log10(no_sc_norm_df[gene])
        low = np.log10(low_sc_norm_df[gene])
        high = np.log10(high_sc_norm_df[gene])
        df[str('no_'+gene)]=no
        df[str('low_'+gene)]=low
        df[str('high_'+gene)]=high
    df = df.replace(to_replace=-np.inf,value=np.nan)
    ax = sns.violinplot(data=df)
    ax.set_ylabel('log2 Counts/cell \n normalized by cell size')
    plt.show()
```
