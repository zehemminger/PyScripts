```python
sham = pd.read_csv('/scratch/sham_m3_slide4_zach_export.csv',index_col=0)
sham
```

```python
plt.hist(np.log10(sham.ave),bins=50)
plt.show()
```

```python
G = []
C = []
for g,c in Counter(sham[sham.ave>10**1.5].gene).items():
    G.append(g)
    C.append(c)
merfish = pd.DataFrame(C,index=G)
```

```python
averages = pd.read_csv('/bigstore/binfo/mouse/Hippocampus/Allen/normalized_cell_type_average_counts.csv.gz',index_col=0)
averages
```

```python
averages.columns
```

```python
shared = list(set(G).intersection(list(averages.index)))
```

```python
X = merfish.loc[shared]
Y = np.mean(averages.loc[shared],axis=1)
plt.scatter(np.log10(X+1),np.log10(Y+1))
plt.xlabel('MERFISH')
plt.ylabel('SCRNAseq')
plt.show()
spearmanr(X,Y)
```

```python
X = merfish.loc[shared]
Y = bulk.loc[shared]
plt.scatter(np.log10(X+1),np.log10(Y+1))
plt.xlabel('MERFISH')
plt.ylabel('SCRNAseq')
plt.show()
spearmanr(X,Y)
```

```python
X = merfish.loc[shared]
corr = []
for ct in averages.columns:
    Y = averages[ct].loc[shared]
    corr.append(spearmanr(X,Y)[0])
plt.hist(corr)
plt.show()
```

```python
X = averages[list(averages.columns)[5]].loc[shared]
Y = averages[list(averages.columns)[0]].loc[shared]
plt.scatter(np.log10(X+1),np.log10(Y+1))
plt.xlabel('SCRNAseq')
plt.ylabel('SCRNAseq')
plt.show()
```
