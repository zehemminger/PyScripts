```python
cells = []
samples = []
cell_types = []
for cell in MeA_AllCells_DGE.columns:
    c,s,t = cell.split('_')
    cells.append(cell)
    samples.append(s)
    cell_types.append(t)
np.unique(cell_types)
info_df = pd.DataFrame(index=cells)
info_df['cells'] = cells
info_df['samples'] = samples
info_df['cell_types'] = cell_types
info_df['barcode'] = [i.split('_')[0] for i in info_df['cells']]
info_df
```

```python
all_cells = len(info_df)
print('All:',all_cells,'cells')
for cell_type in info_df['cell_types'].unique():
    ncells = len(info_df[info_df['cell_types']==cell_type])
    print(cell_type,':',ncells,'cells',round(100*ncells/all_cells,2),'%')
```

```python
Gene_Matrix_Avg = pd.DataFrame(index=MeA_AllCells_DGE.index)
Gene_Matrix_std = pd.DataFrame(index=MeA_AllCells_DGE.index)
for cell_type in np.unique(info_df['cell_types']):
    cells = info_df[info_df['cell_types']==cell_type]['cells']
    cell_type_df = MeA_AllCells_DGE[cells]
    Gene_Matrix_Avg[cell_type] = np.average(cell_type_df,axis=1)
    Gene_Matrix_std[cell_type] = np.std(cell_type_df,axis=1)
Gene_Matrix_Avg.head()
```

```python
Gene_Matrix_std.head()
```

```python
pickle.dump(Gene_Matrix_Avg,open(os.path.join(base_path,'Gene_Matrix_Avg.pkl'),'wb'))
pickle.dump(Gene_Matrix_std,open(os.path.join(base_path,'Gene_Matrix_std.pkl'),'wb'))
```
