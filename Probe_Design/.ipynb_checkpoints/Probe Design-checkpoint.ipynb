```python
import random
from Bio import SeqUtils
from fish_helpers import *
from scipy.signal import lfilter
import tqdm.notebook as tqdm
from Bio import SeqIO
from scipy import sparse
from multiprocessing import Pool
from functools import partial
import sys
```

```python
"""
Parse Transcriptome
transcript identifier (tid) (index)
gene identifier (gene)
sequence (seq)
Convert seq to int (intseq)
for each seed length region:
find each identity(hash)
find if it contains N bases(isvalid)
"""
resourcePath = '/bigstore/binfo/mouse/'
rawTranscriptomeFasta = os.path.join(resourcePath,'mer_transcripts.fa')

tids = []
gids = []
seqs = []
probe_len = 30
with open(rawTranscriptomeFasta) as fasta_file:  # Will close handle cleanly
    for seq_record in tqdm.tqdm(SeqIO.parse(fasta_file, 'fasta')):  # (generator)
        tid,gid = seq_record.description.split(' ')
        gid = gid.split('=')[-1]
        seq = str(seq_record.seq)
        if len(seq)>probe_len:
            tids.append(tid)
            gids.append(gid)
            seqs.append(seq)
transcriptome = pd.DataFrame(index=tids)
transcriptome['gene'] = gids
transcriptome['seq'] = seqs
transcriptome.head()
```

```python
"""
Load Annotation Data
"""
annotation = pd.read_csv('/bigstore/binfo/mouse/mart_export_25Nov2019.txt',sep='\t')
annotation.index = list(annotation['Transcript stable ID'])
"""
Filter Transcriptome to tsl<5
"""
tids = list(transcriptome.index)
TSL = []
for tsl in annotation['Transcript support level (TSL)']:
    try:
        TSL.append(tsl.split(' (')[0])
    except:
        TSL.append('tslNA')
keepers = np.unique(TSL)
keepers = [i for i in keepers if not i=='tslNA']
keepers = [i for i in keepers if not i=='tsl5']
mask = []
for tsl in TSL:
    if tsl in keepers:
        mask.append(True)
    else:
        mask.append(False)
TSL_tids = list(annotation[mask].index)
filtered_tids = list(set(tids).intersection(TSL_tids))
filtered_transcriptome = transcriptome.loc[filtered_tids]
len(filtered_transcriptome)
```

```python
"""
calcualte GC and Tm for all possible probes
find all possible seeds (17 bp regions)
mask probes based on Tm and GC
mask seeds based on masked probes
"""
def calculateProbeMetrics(row,probe_len=30,
                          monovalentSalt=0.3,
                          seedLength=17,
                          probeConc=5e-9,
                          nt2int = {'A':0,'C':1,'G':2,'T':3,'N':np.nan},
                            HT = [0,-7.6,-8.4,-7.8,-7.2,
                                  -8.5,-8.0,-10.6,-7.8,
                                  -8.2,-9.8,-8.0,-8.4,
                                  -7.2,-8.2,-8.5,-7.6],
                            ST = [0,-21.3,-22.4,-21.0,-20.4,
                                  -22.7,-19.9,-27.2,-21.0,
                                  -22.2,-24.4,-19.9,-22.4,
                                  -21.3,-22.2,-22.7,-21.3],
                          tm_max = 76,
                          tm_min = 66,
                          GC_max = 63/100,
                          GC_min = 43/100):

    hashBase = [4**i for i in np.linspace(seedLength-1,0,seedLength-1)]
    tid = row[0]
    gene = row[1]['gene']
    seq = row[1]['seq']
    length = len(seq)
    if length>probe_len:
        # Convert sequence to integers {'A':0,'C':1,'G':2,'T':3,'N':np.nan}
        intSeq = np.array([nt2int[i] for i in seq])
        nnID = (4*intSeq[:-1] + intSeq[1:])+1
        nnID = np.nan_to_num(nnID).astype(int)
        # Calculate Free energy
        dG = np.zeros([2,len(seq)-1])
        dG[0,:] = np.array([HT[i] for i in nnID])
        dG[1,:] = np.array([ST[i] for i in nnID])
        # Calculate Entropy and Enthalpy
        H = lfilter(np.ones([probe_len-1]),1,dG[0,:])[probe_len-1:]
        S = lfilter(np.ones([probe_len-1]),1,dG[1,:])[probe_len-1:]
        fivePrimeAT = (1*(intSeq==0)+1*(intSeq==3))[:-(probe_len)]
        threePrimeAT = (1*(intSeq==0)+1*(intSeq==3))[probe_len:]
        H = H+0.2+(2.2*fivePrimeAT)+(2.2*threePrimeAT)
        S = S-5.7+(6.9*fivePrimeAT)+(6.9*threePrimeAT)
        S = S + 0.368*(probe_len-1)*np.log(monovalentSalt)
        # Calcuate Melting Temp for 30 bp probes
        Tm = (H*1000)/(S+1.9872*np.log(probeConc))-273.15
        # Calculate GC content for 30 bp probes
        gc = 1*((intSeq==1)|(intSeq==2))
        gc = lfilter(np.ones([probe_len-1])/probe_len,1,gc)[probe_len:]
        # convert all possible seedLength seq to unique number (hash)
        h = lfilter(hashBase,1,intSeq)[seedLength-1:]
        # mask seeds that belong to only failed probes
        p_mask = 1*(((1*(Tm>tm_max))+(1*(Tm<tm_min))+(1*(gc>GC_max))+(1*(gc<GC_min)))==0)
        h_mask = 1*(np.array([np.sum(p_mask[i:i+seedLength]) for i in range(len(h))])>0)
#         validBase = np.ones(seedLength)
#         isvalid = 1*(lfilter(validBase,1,1*(intSeq<4))[seedLength:]==1)
        values = [gene,seq,length,intSeq,Tm,gc,h,p_mask,h_mask]
        out = pd.DataFrame(values,index=['gene','seq','length','intseq','Tm','GC','hash','p_mask','h_mask'],columns=[tid]).T
        return out
    else:
        print(tid,'Failed')
        return False
def pfunc(partial_transcriptome):
    Output = []
    for row in partial_transcriptome.iterrows():
        if isinstance(row,tuple):
            Output.append(calculateProbeMetrics(row))
    return Output
Input = []
s = 300
indexes = list(filtered_transcriptome.index)
step = int(len(indexes)/s)+1
for i in range(s):
    partial_transcriptome = filtered_transcriptome.iloc[i*step:(i+1)*step]
    Input.append(partial_transcriptome)
sys.stdout.flush()
Output = []
with Pool(30) as p:
    for out in tqdm.tqdm(p.imap(pfunc,Input),total=len(Input)):
        Output.extend(out)
sys.stdout.flush()
new_transcriptome = pd.concat(Output)
```

```python
"""
Find Homology
"""
def calculateOffTargetScores(row,gene_locations,off_target_vector,off_target_expression_vector):
    idx,info = row
    gene = info['gene']
    isoform_locations = gene_locations[gene]
    h = info['hash']
    hm = info['h_mask']
    isoform_score = []
    off_target_score = []
    unique_desc = 'Inner Loop '+str(idx)
    for i,h in tqdm.tqdm(enumerate(h),total=len(h),desc=unique_desc, leave=False):
        if hm[i]:
            homology_location = np.where(off_target_vector==h)[0]
            isoform_overlap = list(set(isoform_locations).intersection(list(homology_location)))
            isoform_counts = np.sum(off_target_expression_vector[isoform_overlap])
            off_target_counts = np.sum(off_target_expression_vector[homology_location])-isoform_counts
            isoform_score.append(isoform_counts)
            off_target_score.append(off_target_counts)
        else:
            isoform_score.append(-1)
            off_target_score.append(-1)
    return idx,isoform_score,off_target_score

def offTargetWrapper(partial_transcriptome,gene_locations,off_target_vector,off_target_expression_vector):
    idxs = []
    isoform_scores = []
    off_target_scores = []
    unique_desc = 'Inner Loop '+''.join(' ' for i in range(int(10*random.random())))
    for row in tqdm.tqdm(partial_transcriptome.iterrows(),total=len(partial_transcriptome),desc=unique_desc, leave=True):
        idx,isoform_score,off_target_score = calculateOffTargetScores(row,gene_locations,off_target_vector,off_target_expression_vector)
        idxs.append(idx)
        isoform_scores.append(isoform_score)
        off_target_scores.append(off_target_score)
    return idxs,isoform_scores,off_target_scores
idxs = []
isoform_scores = []
off_target_scores = []
step = 1
s = int(len(target_transcriptome)/step)+1
for row in tqdm.tqdm(target_transcriptome.iterrows(),total=len(target_transcriptome),desc='Outer Loop'):
    idx,isoform_score,off_target_score = calculateOffTargetScores(row,
                                                                  gene_locations,
                                                                  off_target_vector,
                                                                  off_target_expression_vector)
    idxs.append(idx)
    isoform_scores.append(isoform_score)
    off_target_scores.append(off_target_score)
scores = pd.DataFrame(index = idxs)
scores['isoform_score'] = isoform_scores
scores['off_target_score'] = off_target_scores
scores

# Input = []
# step = 10
# s = int(len(target_transcriptome)/step)+1
# for i in range(s):
#     Input.append(target_transcriptome.iloc[i*step:(i+1)*step])
# sys.stdout.flush()
# pfunc = partial(offTargetWrapper,
#                 gene_locations=gene_locations,
#                 off_target_vector=off_target_vector,
#                 off_target_expression_vector=off_target_expression_vector)
# print('input and pfunc generated')
# idxs = []
# isoform_scores = []
# off_target_scores = []
# with Pool(10) as p:
#     for idx,isoform_score,off_target_score in tqdm.tqdm(p.imap(pfunc,Input),total=len(Input),desc='Outer Loop'):
#         idxs.extend(idx)
#         isoform_scores.extend(isoform_score)
#         off_target_scores.extend(off_target_score)
# scores = pd.DataFrame(index = idxs)
# scores['isoform_score'] = isoform_scores
# scores['off_target_score'] = off_target_scores
# sys.stdout.flush()
# scores
```

```python
import numpy as np
def calculateMaxNumProbes(row,overlap=15,probe_len=30):
    loc = np.where(row[1]['p_mask']==1)[0]
    overlap = probe_len-overlap
    keepers = []
    for probe in loc:
        possible = list(range(probe-overlap,probe+overlap))
        if len(set(possible).intersection(keepers))==0:
            keepers.append(probe)
    nP = len(keepers)
    return row[0],nP
Input = []
for row in tqdm.tqdm(target_transcriptome.iterrows(),total=len(target_transcriptome)):
    Input.append(row)
sys.stdout.flush()
idxs = []
nps = []
with Pool(30) as p:
    for idx,nP in tqdm.tqdm(p.imap(calculateMaxNumProbes,Input),total=len(Input)):
        idxs.append(idx)
        nps.append(nP)
scores = pd.DataFrame(index=idxs)
scores['nps'] = nps
sys.stdout.flush()
```

```python
"""
Build penalties tables for each nonspecific event:
isoforms
other genes
ncRNA
forbidden sequences
"""
def generatePenalties(transcriptome):
    tid = list(transcriptome.index)[0]
    t = transcriptome['hash'].loc[tid]
    transcriptome = transcriptome.drop(index=tid)
    penalties = np.zeros((len(transcriptome),len(t)))
    for i,idx in enumerate(transcriptome.index):
        h = transcriptome['hash'].loc[idx]
        penalties[i,:] = 1*np.array([np.any(h == i) for i in t])
    penalties = sparse.csr_matrix(penalties)
    return tid,penalties

def generateLocalTranscriptome(tid,transcriptome,iso=True):
    target = pd.DataFrame(transcriptome.loc[tid]).T
    gene = transcriptome['gene'].loc[tid]
    if iso:
        localTranscriptome = transcriptome[transcriptome['gene']==gene]
        localTranscriptome = localTranscriptome.drop(index=tid)
    else:
        localTranscriptome = transcriptome[transcriptome['gene']!=gene]
    localTranscriptome = pd.concat([target,localTranscriptome])
    return localTranscriptome
```

```python
"""
Calculate Penalty
"""
def calculate_iso_penalties(localTranscriptome):
    out = {}
    for tid in localTranscriptome.index:
        t = localTranscriptome['hash'].loc[tid]
        if len(localTranscriptome)>0:
            penalties = np.zeros((len(localTranscriptome),len(t)))
            for i,idx in enumerate(localTranscriptome.index):
                if idx!=tid:
                    h =  localTranscriptome['hash'].loc[idx]
                    e = localTranscriptome['FPKM'].loc[idx]
                    penalties[i,:] = e*1*np.array([np.any(h == i) for i in t])
            penalties = np.sum(penalties,axis=0)
        else:
            penalties = np.zeros_like(t)
        out[tid] = penalties
    return out
Input = []
for gene in tqdm.tqdm(slicedTranscriptome['gene'].unique()):
    Input.append(slicedTranscriptome[slicedTranscriptome['gene']==gene])    
tids = []
iso_penalties = []
sys.stdout.flush()
with Pool(30) as p:
    for tid_dict in tqdm.tqdm(p.imap(calculate_iso_penalties,Input,chunksize=1),total=len(Input)):
        for tid,iso_penaltie in tid_dict.items():
            tids.append(tid)
            iso_penalties.append(iso_penaltie)
sys.stdout.flush()
iso_penalties_df = pd.DataFrame(iso_penalties,index=tids)
iso_penalties_df.head()
```
