```python
# Parse Fasta
f = '/bigstore/binfo/zebra_finch/NCBI/GCF_003957565.1_bTaeGut1_v1.p_transcripts.fa'
from Bio import SeqIO
with open(f) as fasta_file:  # Will close handle cleanly
    gid = []
    seqs = []
    records = []
    for seq_record in SeqIO.parse(fasta_file, 'fasta'):  # (generator)
        gid.append(seq_record.id)
        seqs.append(str(seq_record.seq))
        records.append(seq_record)
fasta = pd.DataFrame()
fasta['id'] = gid
fasta['sequence'] = seqs
fasta['record'] = records
fasta
```
