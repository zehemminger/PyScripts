```python
seq = probes.iloc[0]['sequence']
def reverse_complement(seq):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'} 
    rev = seq[::-1]
    new_seq = []
    for s in rev:
        new_seq.append(complement[s.upper()])
    return ''.join(new_seq)
print(seq)
print(reverse_complement(seq))
```
