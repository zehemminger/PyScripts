```python
expression_worksheet = client.open("MAPK_genes").sheet1
expression = expression_worksheet.get_all_records()
expression_df = pd.DataFrame(expression)

Transcript_IDs = expression_worksheet.range('G2:G'+str(len(expression_df)+1))
Gene_IDs = expression_worksheet.range('F2:F'+str(len(expression_df)+1))
FPKMs = expression_worksheet.range('C2:C'+str(len(expression_df)+1))
lengths = expression_worksheet.range('D2:D'+str(len(expression_df)+1))
Descriptions = expression_worksheet.range('B2:B'+str(len(expression_df)+1))
Genes = expression_worksheet.range('A2:A'+str(len(expression_df)+1))
Isoforms = expression_worksheet.range('I2:I'+str(len(expression_df)+1))
Type = expression_worksheet.range('J2:J'+str(len(expression_df)+1))

f = '/bigstore/binfo/mouse/3T3/GSM2609311_cufflinks/isoforms.fpkm_tracking'
ReadsPerGene = pd.read_csv(f,sep='\t')
ReadsPerGene.index=ReadsPerGene.tracking_id
fpkm = ReadsPerGene

annotation = pd.read_csv('/bigstore/binfo/mouse/mart_export_25Nov2019.txt',sep='\t')

flag = 'gene'
FPKM_gene_name_col = 'gene_short_name'#'Gene name'
FPKM_gid_col = 'gene_id'#'Gene stable ID'
FPKM_tid_col = 'tracking_id'#'Transcript stable ID'
tid_col = 'Transcript stable ID'
gid_col = 'Gene stable ID'
ttype_col = 'Transcript type'
gene_name_col = 'Gene name'
length_col = 'Transcript length (including UTRs and CDS)'
gene_description_col = 'Gene description'
FPKM_FPKM_col = 'FPKM'#'IFNAR-0_tot.bam'
completed = []

for i,cell in enumerate(Genes):
    gene = cell.value
    if gene == '':
        continue
    if gene in completed:
        Type[i].value= 'duplicate'
        print(gene,' duplicated')
    if flag=='gene':
        try:
            gid = fpkm[fpkm[FPKM_gene_name_col]==gene][FPKM_gid_col].unique()[0]
        except:
            gid = 'None'
    elif flag == 'gid':
        gid = gene
    elif flag == 'tid':
        try:
            gid = fpkm[fpkm[FPKM_tid_col]==gene][FPKM_gid_col].unique()[0]
            master_tid = gene.split('.')[0]
        except:
            gid = 'None'
    else:
        print('flag is wrong')
    if gid=='None':
        Type[i].value= 'Failed'
        Transcript_IDs[i].value = ''
        Gene_IDs[i].value = ''
        tid_fpkm = ''
        FPKMs[i].value = ''
        lengths[i].value = ''
        Descriptions[i].value = ''
#         Genes[i].value = ''
        print('gid not found for ',gene,gid)
        continue
    FPKM = []
    I = []
    g_annotation = annotation[annotation[gid_col]==gid]
    g_annotation.drop_duplicates(tid_col)
    for t in g_annotation[tid_col]:
        try:
            f = fpkm[fpkm[FPKM_tid_col]==t][FPKM_FPKM_col].iloc[0]
            g = np.sum(fpkm[fpkm[FPKM_gid_col]==gid][FPKM_FPKM_col])
            FPKM.append(f)
            I.append(f/g)
        except:
            FPKM.append(np.nan)
            I.append(np.nan)
    g_annotation['FPKM'] = FPKM
    g_annotation['I'] = I
    gene_FPKM_df = g_annotation
    if 'protein_coding' in list(gene_FPKM_df['Transcript type']):
        gene_FPKM_df = gene_FPKM_df[gene_FPKM_df['Transcript type']=='protein_coding']
    mask = []
    for t in gene_FPKM_df['Transcript support level (TSL)']:
        try:
            if 'tsl1' in t:
                mask.append(True)
            else: 
                mask.append(False)
        except:
            mask.append(False)
    if len(gene_FPKM_df[mask])>0:
        gene_FPKM_df = gene_FPKM_df[mask]
    
    gene_FPKM_df = gene_FPKM_df.sort_values(length_col, ascending=False)
    gene_FPKM_df = gene_FPKM_df.sort_values('FPKM', ascending=False)
    
    if len(gene_FPKM_df)==0:
        print('Failed finding: ', gid)
        Type[i].value= 'Failed'
        Transcript_IDs[i].value = ''
        Gene_IDs[i].value = ''
        tid_fpkm = ''
        FPKMs[i].value = ''
        lengths[i].value = ''
        Descriptions[i].value = ''
#         Genes[i].value = ''
        continue
    tid = str(gene_FPKM_df[tid_col])
    tid_FPKM_df = gene_FPKM_df.iloc[0]
    Transcript_IDs[i].value = str(tid_FPKM_df[tid_col])
    Gene_IDs[i].value = str(tid_FPKM_df[gid_col])
    tid_fpkm = tid_FPKM_df['FPKM']
    FPKMs[i].value = tid_FPKM_df['FPKM']
    lengths[i].value = int(tid_FPKM_df[length_col])
    Descriptions[i].value = str(tid_FPKM_df['WikiGene description'])
    Genes[i].value = str(tid_FPKM_df[gene_name_col])
    Type[i].value= 'Passed'
    if tid_fpkm==0:
        I = 0
    else:
        I = tid_FPKM_df['FPKM']/np.sum(g_annotation['FPKM'])
    Isoforms[i].value = I
    completed.append(gene)
expression_worksheet.update_cells(Transcript_IDs)
expression_worksheet.update_cells(FPKMs)
expression_worksheet.update_cells(lengths)
#expression_worksheet.update_cells(Genes)
expression_worksheet.update_cells(Gene_IDs)
expression_worksheet.update_cells(Isoforms)
expression_worksheet.update_cells(Type)
expression_worksheet.update_cells(Descriptions)
```
